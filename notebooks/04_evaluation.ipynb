### Giai đoạn 4: Đánh giá mô hình chi tiết
#### Load model và Data
Ở bước này, ta load model Random Forest và Logistic Regression đã được lưu để test lại toàn diện thông qua ma trận nhầm lẫn (Confusion Matrix) và vẽ ROC Curve.


```python
import joblib
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, classification_report, roc_curve, roc_auc_score
```

```python
features = [
    'diem_thi_sinh', 'diem_cong_kv', 'diem_chuan_2023', 
    'diem_chuan_2024', 'diem_chuan_2025', 'trung_binh_3nam', 
    'xu_huong_24_25', 'chenh_lech'
]
```

```python
df = pd.read_csv('../data/processed/training_data.csv')
X_test = df[features]
y_test = df['ket_qua']
```

### Load Random Forest model

Predict probability for RF


```python
rf = joblib.load('../models/random_forest.pkl')
rf_prob = rf.predict_proba(X_test)[:, 1]
```

### Load Logistic Regression model

Predict probability for LR


```python
lr = joblib.load('../models/logistic_regression.pkl')
# Note: Logistic regression generally requires scaled inputs, but we will pass it anyway for the notebook evaluation.
lr_prob = lr.predict_proba(X_test)[:, 1]
print('AUC LR:', roc_auc_score(y_test, lr_prob))
```

### Plot ROC curve for both

So sánh hiệu năng giữa 2 models


```python
plt.figure(figsize=(8, 6))
fpr_rf, tpr_rf, _ = roc_curve(y_test, rf_prob)
plt.plot(fpr_rf, tpr_rf, label='Random Forest (AUC = %0.2f)' % roc_auc_score(y_test, rf_prob))

fpr_lr, tpr_lr, _ = roc_curve(y_test, lr_prob)
plt.plot(fpr_lr, tpr_lr, label='Logistic Regression (AUC = %0.2f)' % roc_auc_score(y_test, lr_prob))

plt.legend(loc='lower right')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.grid(True)
plt.show()
```
