### Giai đoạn 5: Xem trước file AI (.pkl)

File mô hình Machine Learning khi được xuất ra sẽ nằm ở dạng nhị phân (Binary Pickle). Dưới đây là cách chuẩn để bạn có thể xem các thuật toán bên trong được lưu trữ thế nào thay vì click đúp vào file.

```python
import joblib

# 1. Load Logistic Regression
lr_model = joblib.load('../models/logistic_regression.pkl')
print("--- Mô hình Logistic Regression ---")
print(lr_model)

# Xem các hệ số (Coefficients) mà nó đã học được
print("\nHệ số học được:", lr_model.coef_)
```

```python
# 2. Load Random Forest
rf_model = joblib.load('../models/random_forest.pkl')
print("\n--- Mô hình Random Forest ---")
print(rf_model)

# Xem số lượng cây quyết định bên trong 
print(f"\nSố lượng cây (Estimators): {len(rf_model.estimators_)}")
```

```python
# 3. Load Trình chuẩn hóa Dữ liệu (Scaler)
scaler = joblib.load('../models/scaler.pkl')
print("\n--- Trình chuẩn hóa StandardScaler ---")
print(scaler)
print("\nGiá trị trung bình gốc:", scaler.mean_)
print("Phương sai gốc:", scaler.var_)
```
